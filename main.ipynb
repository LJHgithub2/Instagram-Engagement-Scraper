```python
# 라이브러리 임포트
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains
from dotenv import load_dotenv
import os
import urllib.request
import ssl
import time
# 크롬 드라이버 자동 업데이트
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# .env 파일에서 환경 변수 로드
load_dotenv()

ssl._create_default_https_context = ssl._create_unverified_context


# 브라우저 꺼짐 방지
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)

# 불필요한 에러 메시지 없애기
chrome_options.add_experimental_option("excludeSwitches", ["enable-logging"])
service = Service(executable_path=ChromeDriverManager().install())
```

```python

driver = webdriver.Chrome(options=chrome_options)

# 웹페이지 해당 주소 이동
driver.get("https://instagram.com")


username = os.getenv("INSTAGRAM_USERNAME")
password = os.getenv("INSTAGRAM_PASSWORD")
user = os.getenv("INSTAGRAM_ID")


#인스타그램 자동로그인
driver.implicitly_wait(10)  #나타날때까지 최대 10초동안 기다려
login_id = driver.find_element(By.CSS_SELECTOR, 'input[name="username"]')
login_id.send_keys(username) # 아이디 입력
login_pwd = driver.find_element(By.CSS_SELECTOR, 'input[name="password"]')
login_pwd.send_keys(password) # 비번 입력
driver.implicitly_wait(10)
login_id.send_keys(Keys.ENTER) #enter 키를 쳐주세요
time.sleep(5)


driver.get(f"https://www.instagram.com/{user}/")
driver.implicitly_wait(15)

# 추가 게시물 확인 및 데이터 로딩
time.sleep(2)
driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
time.sleep(2)
    


followers_count_element = driver.find_element(By.XPATH, '/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[2]/div[2]/section/main/div/header/section/ul/li[2]/a/span/span')
print("팔로우 수: " + followers_count_element.text)

#첫번째 사진 누름
post = driver.find_elements(By.CSS_SELECTOR, '._aagw')
driver.implicitly_wait(15)

'''
좋아요수와 댓글수가 동일한 선택자로 구성됨
두 값을 같이 받아온 후, 슬라이싱
'''

for i in range(min(len(post), 20)):
    print(f"--------------{i+1} 게시물-----------------")

    #게시물로 마우스 옮기기 실행
    act = ActionChains(driver)
    act.move_to_element(post[i]).perform()
    # 크롬 드라이버 종료

    likes = driver.find_element(By.XPATH,f'/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[2]/div[2]/section/main/div/div[3]/div/div[{int(i/3)+1}]/div[{i%3 + 1}]/a/div[3]/ul/li[1]/span[1]')
    commnets = driver.find_element(By.XPATH,f'/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[2]/div[2]/section/main/div/div[3]/div/div[{int(i/3)+1}]/div[{i%3 + 1}]/a/div[3]/ul/li[2]/span[1]')
    #인덱스를 사용해 좋아요수와 댓글수 구분
    likes = likes.text
    comments = commnets.text

    print("좋아요 수 : " +likes)
    print("댓글 수 : " +comments)


driver.quit()
```
